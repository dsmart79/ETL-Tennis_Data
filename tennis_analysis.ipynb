```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
match_stats_csv = "tennis_stats.csv"
match_stats_df = pd.read_csv(match_stats_csv)
match_stats_df.head()
```

```python
new_match_stats_df = match_stats_df[['match_id', 'winner_first_serve_points_won', 'winner_first_serve_points_total', 'winner_second_serve_points_won', 'winner_second_serve_points_total', 'loser_first_serve_points_won', 'loser_first_serve_points_total','loser_second_serve_points_won', 'loser_second_serve_points_total', 'winner_break_points_converted', 'winner_break_points_return_total', 'loser_break_points_converted', 'loser_break_points_return_total']].copy()
new_match_stats_df.head()
```

```python
winner_first_serve_percentage = []
for x in range(len(new_match_stats_df.winner_first_serve_points_won)):
    winner_first_serve_percentage.append(new_match_stats_df.winner_first_serve_points_won[x]/new_match_stats_df.winner_first_serve_points_total[x])
new_match_stats_df['winner_first_serve_percentage'] = winner_first_serve_percentage
#new_match_stats_df.head()
```

```python
winner_second_serve_percentage = []
for x in range(len(new_match_stats_df.winner_first_serve_points_won)):
    winner_second_serve_percentage.append(new_match_stats_df.winner_second_serve_points_won[x]/new_match_stats_df.winner_second_serve_points_total[x])
new_match_stats_df['winner_second_serve_percentage'] = winner_second_serve_percentage
new_match_stats_df.head()
```

```python
loser_first_serve_percentage = []
for x in range(len(new_match_stats_df.loser_first_serve_points_won)):
    loser_first_serve_percentage.append(new_match_stats_df.loser_first_serve_points_won[x]/new_match_stats_df.loser_first_serve_points_total[x])
new_match_stats_df['loser_first_serve_percentage'] = loser_first_serve_percentage
#new_match_stats_df.head()
```

```python
loser_second_serve_percentage = []
for x in range(len(new_match_stats_df.loser_first_serve_points_won)):
    loser_second_serve_percentage.append(new_match_stats_df.loser_second_serve_points_won[x]/new_match_stats_df.loser_second_serve_points_total[x])
new_match_stats_df['loser_second_serve_percentage'] = loser_second_serve_percentage
new_match_stats_df.head()
```

```python
winner_break_point_percentage = []
for x in range(len(new_match_stats_df.loser_first_serve_points_won)):
    winner_break_point_percentage.append(new_match_stats_df.winner_break_points_converted[x]/new_match_stats_df.winner_break_points_return_total[x])
new_match_stats_df['winner_break_point_percentage'] = winner_break_point_percentage
new_match_stats_df.head()
```

```python
loser_break_point_percentage = []
for x in range(len(new_match_stats_df.loser_first_serve_points_won)):
    loser_break_point_percentage.append(new_match_stats_df.loser_break_points_converted[x]/new_match_stats_df.loser_break_points_return_total[x])
new_match_stats_df['loser_break_point_percentage'] = loser_break_point_percentage
new_match_stats_df.columns.values
```

```python
statsdf = new_match_stats_df[['match_id', 'winner_first_serve_points_won','winner_first_serve_points_total',
                              'winner_first_serve_percentage','winner_second_serve_points_won','winner_second_serve_points_total','winner_second_serve_percentage',
                              'loser_first_serve_points_won','loser_first_serve_points_total','loser_first_serve_percentage', 'loser_second_serve_points_won','loser_second_serve_points_total', 'loser_second_serve_percentage',
                              'winner_break_points_converted','winner_break_points_return_total',
                              'winner_break_point_percentage','loser_break_points_converted','loser_break_points_return_total',
                              'loser_break_point_percentage']]

statsdf['winner_first_serve_times100'] = statsdf['winner_first_serve_percentage']*100
statsdf['loser_first_serve_times100'] = statsdf['loser_first_serve_percentage']*100
statsdf['winner_second_serve_times100'] = statsdf['winner_second_serve_percentage']*100
statsdf['loser_second_serve_times100'] = statsdf['loser_second_serve_percentage']*100
statsdf['winner_bpp_times100'] = statsdf['winner_break_point_percentage']*100
statsdf['loser_bpp_times100'] = statsdf['loser_break_point_percentage']*100
statsdf.drop(columns=['winner_first_serve_percentage','loser_first_serve_percentage','winner_break_point_percentage',
                      'loser_break_point_percentage','winner_second_serve_percentage','loser_second_serve_percentage'], axis =1, inplace=True)

statsdf = statsdf.round(1)

statsdf = statsdf.astype(str)
#statsdf.head()
#statsdf.dtypes

statsdf['winner_first_serve_summary'] = statsdf['winner_first_serve_times100'] + '%' + ' ' + '(' + statsdf['winner_first_serve_points_won'] + '/' + statsdf['winner_first_serve_points_total'] + ')'
statsdf['loser_first_serve_summary'] = statsdf['loser_first_serve_times100'] + '%' + ' ' + '(' + statsdf['loser_first_serve_points_won'] + '/' + statsdf['loser_first_serve_points_total'] + ')'
statsdf['winner_second_serve_summary'] = statsdf['winner_second_serve_times100'] + '%' + ' ' + '(' + statsdf['winner_second_serve_points_won'] + '/' + statsdf['winner_second_serve_points_total'] + ')'
statsdf['loser_second_serve_summary'] = statsdf['loser_second_serve_times100'] + '%' + ' ' + '(' + statsdf['loser_second_serve_points_won'] + '/' + statsdf['loser_second_serve_points_total'] + ')'
statsdf['winner_break_points_summary'] = statsdf['winner_bpp_times100'] + '%' + ' ' + '(' + statsdf['winner_break_points_converted'] + '/' + statsdf['winner_break_points_return_total'] + ')'
statsdf['loser_break_points_summary'] = statsdf['loser_bpp_times100'] + '%' + ' ' + '(' + statsdf['loser_break_points_converted'] + '/' + statsdf['loser_break_points_return_total'] + ')'


statsdf.drop(columns=['winner_first_serve_points_won',
       'winner_first_serve_points_total', 'loser_first_serve_points_won',
       'loser_first_serve_points_total', 'winner_second_serve_points_won',
       'winner_second_serve_points_total', 'loser_second_serve_points_won',
       'loser_second_serve_points_total','winner_break_points_converted',
       'winner_break_points_return_total', 'loser_break_points_converted',
       'loser_break_points_return_total', 'winner_first_serve_times100',
       'loser_first_serve_times100','winner_second_serve_times100',
       'loser_second_serve_times100', 'winner_bpp_times100',
       'loser_bpp_times100'], axis = 1, inplace = True)

statsdf.head()
```

```python
match_scores_csv = "tennis_scores.csv"
match_scores_df = pd.read_csv(match_scores_csv)
match_scores_df.head()
```

```python
new_match_scores_df = match_scores_df[['tourney_year_id','tourney_slug', 'match_id', 'winner_name', 'loser_name', 'match_score_tiebreaks']]
new_match_scores_df.rename(columns = {'match_score_tiebreaks' : 'match_score', 'tourney_slug':'tourney_name'}, inplace=True)
new = new_match_scores_df["tourney_year_id"].str.split("-", n = 1, expand = True)
new_match_scores_df["tourney_year"]= new[0]
new_match_scores_df["id"]= new[1]
new_match_scores_df.drop(['tourney_year_id', 'id'], axis = 1, inplace= True)
new_match_scores_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/ETL_Tennis"
tennis_engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
tennis_engine.table_names()
```

```python
statsdf.to_sql(name='tennis_statistics', con=tennis_engine, if_exists='append', index=False)
```

```python
new_match_scores_df.to_sql(name='tennis_scores', con=tennis_engine, if_exists='append', index=False)
```
